```python
import numpy as np
import tensorflow as tf
import tensorflow.keras as keras
# import keras
# from keras import backend as K
from tensorflow.keras import backend as K
from tensorflow.keras.layers import Input, Conv2D, AveragePooling2D, UpSampling2D, Concatenate, concatenate
import os
import skimage
import skimage.io
import skimage.transform
import glob
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def process_voc(mask_dir, target_height, target_width):
    files = glob.glob(mask_dir + '/*.png')
    for file in files:
        img_mask = skimage.io.imread(file)
        if img_mask.shape[0] != target_height or img_mask.shape[1] != target_width:
            img_mask = skimage.transform.resize(img_mask, (target_height, target_width), order=0) # Resize.
        img_mask = (img_mask * 255).astype(np.int32)
        img_mask_bin = convert_mask(img_mask)
        np.save(file.replace('.png', '.npy'), img_mask_bin)

def color_map(N=256, normalized=False):
    def bitget(byteval, idx):
        return ((byteval & (1 << idx)) != 0)

    dtype = 'float32' if normalized else 'uint8'
    cmap = np.zeros((N, 3), dtype=dtype)
    for i in range(N):
        r = g = b = 0
        c = i
        for j in range(8):
            r = r | (bitget(c, 0) << 7-j)
            g = g | (bitget(c, 1) << 7-j)
            b = b | (bitget(c, 2) << 7-j)
            c = c >> 3

        cmap[i] = np.array([r, g, b])

    cmap = cmap / 255 if normalized else cmap
    return cmap

def convert_mask(mask):
    # Takes colorized input mask of shape [H, W, 3] and returns true mask of shape [H, W, 1].
    bin_mask = np.zeros(shape=(mask.shape[0], mask.shape[1], 1), dtype=np.int32)
    for h in range(mask.shape[0]):
        for w in range(mask.shape[1]):
            cval = tuple(mask[h][w])
            if not cval in mapping_by_color:
                # Ignore, unlabeled/border pixel.
                target = -1
            else:
                class_name = mapping_by_color[cval]
                target = labels.index(class_name)
            bin_mask[h, w] = target
    return bin_mask
```

```python
class CustomDataGenerator(keras.utils.Sequence):
    def __init__(self, source_raw, source_mask, filenames, batch_size, target_height, target_width):
        self.source_raw = source_raw
        self.source_mask = source_mask
        self.filenames = filenames
        self.batch_size = batch_size
        self.target_height = target_height
        self.target_width = target_width
        self.on_epoch_end()

    def on_epoch_end(self):
        '''Shuffle list of files after each epoch.'''
        np.random.shuffle(self.filenames)
        
    def __getitem__(self, index):
        cur_files = self.filenames[index*self.batch_size:(index+1)*self.batch_size]
        # Generate data
        X, y = self.__data_generation(cur_files)
        return X, y
    
    def __data_generation(self, cur_files):
        X = np.empty(shape=(self.batch_size, self.target_height, self.target_width, 3))
        Y = np.empty(shape=(self.batch_size, self.target_height, self.target_width, 1), dtype=np.int32)
        
        for i, file in enumerate(cur_files):
            img_raw = skimage.io.imread(os.path.join(self.source_raw, file) + '.jpg')
            if img_raw.shape[0] != self.target_height or img_raw.shape[1] != self.target_width:
                img_raw = skimage.transform.resize(img_raw, (self.target_height, self.target_width)) # Resize.
            img_raw = img_raw.astype(np.float32) * 255.  # [0, 255].
            print('in:', img_raw.min(), img_raw.mean(), img_raw.max())
            
            
            # TODO(casser): DOUBLE-CHECK CORRECTNESS. SHOULD INPUT BE NORMALIZED ALREADY?
            X[i] = keras.applications.resnet50.preprocess_input(img_raw)
            
            # X[i] = img_raw * 2.0 - 1.0  # see TF mode in imagenet_utils.py for MobileNet; scale to [-1, 1].            
            Y[i] = np.load(os.path.join(self.source_mask, file) + '.npy') 
            
            # img_mask = skimage.io.imread(os.path.join(self.source_mask, file) + '.png')
            # if img_mask.shape[0] != self.target_height or img_mask.shape[1] != self.target_width:
            #     img_mask = skimage.transform.resize(img_mask, (self.target_height, self.target_width), order=0) # Resize.
            # img_mask = (img_mask * 255).astype(np.int32)
        return X, Y
    
    def __len__(self):
        return int(np.floor(len(self.filenames) / self.batch_size))
```

```python
gen_train = CustomDataGenerator(source_raw=source_raw,
                                source_mask=source_mask,
                                filenames=files_train,
                                batch_size=8,
                                target_height=128,
                                target_width=128)
```

```python
K.clear_session()

def get_unet_resnet():
    def conv_block_simple(prev, num_filters, name):
        return Conv2D(num_filters, kernel_size=(3, 3), padding='same', name=name)(prev)
    
    resnet_encoder = keras.applications.resnet50.ResNet50(input_shape=(224, 224, 3), 
                                                          include_top=False,
                                                          weights='imagenet')

    for l in resnet_encoder.layers:
        # print(l.name)
        l.trainable = True
    conv0 = resnet_encoder.get_layer("activation").output
    conv1 = resnet_encoder.get_layer("activation_1").output
    conv2 = resnet_encoder.get_layer("activation_10").output
    conv3 = resnet_encoder.get_layer("activation_22").output
    conv4 = resnet_encoder.get_layer("activation_40").output
    # conv5 = resnet_encoder.get_layer("activation_48").output
    
    print('conv0', conv0)
    print('conv1', conv1)
    print('conv2', conv2)
    print('conv3', conv3)
    print('conv4', conv4)
    # print('conv5', conv5)
    

    up6 = conv4 #concatenate([UpSampling2D()(conv5), conv4], axis=-1)
    conv6 = conv_block_simple(up6, 256, "conv6_1")
    conv6 = conv_block_simple(conv6, 256, "conv6_2")

    up7 = concatenate([UpSampling2D()(conv6), conv3], axis=-1)
    conv7 = conv_block_simple(up7, 192, "conv7_1")
    conv7 = conv_block_simple(conv7, 192, "conv7_2")

    up8 = concatenate([UpSampling2D()(conv7), conv2], axis=-1)
    conv8 = conv_block_simple(up8, 128, "conv8_1")
    conv8 = conv_block_simple(conv8, 128, "conv8_2")

    up9 = concatenate([UpSampling2D()(conv8), conv1], axis=-1)
    conv9 = conv_block_simple(up9, 64, "conv9_1")
    conv9 = conv_block_simple(conv9, 64, "conv9_2")

    up10 = concatenate([UpSampling2D()(conv9), conv0], axis=-1)
    conv10 = conv_block_simple(up10, 32, "conv10_1")
    conv10 = conv_block_simple(conv10, 32, "conv10_2")
    
    up11 = UpSampling2D()(conv10)
    conv11 = conv_block_simple(up11, 32, "conv11_1")
    conv11 = conv_block_simple(conv11, 32, "conv11_2")

    x = Conv2D(nclasses, (1, 1), activation="softmax", name="prediction")(conv11)
    
    model = tf.keras.Model(resnet_encoder.input, x)
    return model

model = get_unet_resnet()
```

```python
def masked_loss(y_true, y_pred):
    # y_true: [B, H, W, 1].
    # y_pred: [B, H, W, nclasses].
    gt_validity_mask = tf.cast(tf.greater_equal(y_true[:, :, :, 0], 0), dtype=tf.float32) # [B, H, W]
    
    # keras sparse categorical crossentropy loss expects labels >= 0.
    y_true = K.abs(y_true)
    
    # y_true = K.sign(y_true) # Test for pure foreground-background segmentation.
    
    raw_loss = keras.losses.sparse_categorical_crossentropy(y_true, y_pred)  # [B, H, W].
    masked = gt_validity_mask * raw_loss
    return tf.reduce_mean(masked)
```

```python
model.compile(optimizer=keras.optimizers.Adam(lr=0.0005), loss=masked_loss)
```
